```python
# Calculate Indicators using pandas_ta
bbands = ta.bbands(data_1min['close'], length=20, std=2)

# Assign the Bollinger Bands to the DataFrame
data_1min['BB_upper'] = bbands['BBU_20_2.0']
data_1min['BB_middle'] = bbands['BBM_20_2.0']
data_1min['BB_lower'] = bbands['BBL_20_2.0']
data_1min['RSI'] = ta.rsi(data_1min['close'], length=14)
data_1min['EMA_50'] = ta.ema(data_1min['close'], length=50)
data_1min['EMA_200'] = ta.ema(data_1min['close'], length=200)

# Calculate MACD
macd = ta.macd(data_1min['close'])

# Assign the correct MACD columns
data_1min['MACD'] = macd['MACD_12_26_9']  # Update based on actual column name
data_1min['MACD_Signal'] = macd['MACDs_12_26_9']  # Update based on actual column name

# Handle NaN values
data_1min.dropna(subset=['MACD', 'MACD_Signal'], inplace=True)  # Drop rows with NaN MACD values

# Calculate VWAP
data_1min['VWAP'] = ta.vwap(data_1min['high'], data_1min['low'], data_1min['close'], data_1min['volume'])  # Calculate VWAP
```

```python
# Define a function to calculate the confidence score and generate signals
def mix_strategy(price, bb_upper, bb_lower, rsi, macd, macd_signal):
    score = 0
    
    # Condition 1: Price below BB lower (Buy signal)
    if price < bb_lower:
        score += 1
    
    # Condition 2: Price above BB upper (Sell signal)
    if price > bb_upper:
        score -= 1
    
    # Condition 3: RSI < 30 (Buy signal)
    if rsi < 30:
        score += 1
    
    # Condition 4: RSI > 70 (Sell signal)
    if rsi > 70:
        score -= 1
    
    # Condition 5: MACD above MACD signal (Buy signal)
    if macd > macd_signal:
        score += 1
    
    # Condition 6: MACD below MACD signal (Sell signal)
    if macd < macd_signal:
        score -= 1
    
    # Generate signal based on score
    if score >= 4:
        return "Buy"
    elif score == 3:
        return "Partial Buy"
    elif score == 2:
        return "Hold"
    elif score == 1:
        return "Partial Sell"
    else:
        return "Sell"

# Apply the mixed strategy to the entire dataset
data_1min['Signal'] = data_1min.apply(
    lambda row: mix_strategy(row['close'], row['BB_upper'], row['BB_lower'], row['RSI'], row['MACD'], row['MACD_Signal']),
    axis=1
)

# Now the 'Signal' column will contain the mix strategy signals
print(data_1min[['close', 'BB_upper', 'BB_lower', 'RSI', 'MACD', 'MACD_Signal', 'Signal']])
```

```python
def calculate_pnl(data):
    initial_capital = 100000  # Starting capital: $100,000
    position_size = 1000  # Units to trade per signal
    capital = initial_capital
    position = 0  # Current number of units owned
    entry_price = 0  # Entry price for the position
    pnl = []  # Track PnL for each row
    
    # Track whether PnL should be added or not (to ensure alignment)
    in_position = False
    pnl.append(0)
    pnl.append(0)
    # Skip the first 2 days as requested
    for i in range(2, len(data)):
        row = data.iloc[i]
        signal = row['Signal']
        price = row['close']

        # Buy or Partial Buy: Open/increase a position
        if signal == 'Buy' or signal == 'Partial Buy':
            if position == 0:  # If no position, open a new one
                position = position_size
                entry_price = price
                in_position = True
            elif position > 0:  # Already in a position, increase size
                position += position_size
                entry_price = (entry_price * (position - position_size) + price * position_size) / position
            pnl.append(0)  # No PnL yet until the position is closed

        # Sell or Partial Sell: Close/reduce the long position
        elif signal == 'Sell' or signal == 'Partial Sell':
            if position > 0:  # Only if we are holding a position
                # Calculate profit/loss for current position
                trade_pnl = (price - entry_price) * position
                capital += trade_pnl
                pnl.append(trade_pnl)
                
                # Close the position fully or partially
                if signal == 'Sell':
                    position = 0  # Close all positions
                    in_position = False
                elif signal == 'Partial Sell':
                    position -= position_size  # Reduce position
                    if position <= 0:
                        position = 0
                        in_position = False
            else:
                pnl.append(0)

        # Hold: No trading, no change in position
        else:
            pnl.append(0)  # No change in PnL for 'Hold'

    return pnl, capital

# Apply the PnL calculation, skipping the first 2 days
pnl, final_capital = calculate_pnl(data_1min.iloc[2:])
data_1min['PnL'] = [0] * 2 + pnl  # PnL starts from the 3rd row

# Display final capital and total profit/loss
print("Final capital:", final_capital)
print("Total profit/loss:", final_capital - 100000)

# Display the last few rows of the data with PnL
print(data_1min[['close', 'Signal', 'PnL']].tail(10))
```
